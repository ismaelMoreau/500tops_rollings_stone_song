```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

url = 'https://www.rollingstone.com/music/music-lists/best-songs-of-all-time-1224767/'


# Start a new instance of Chrome
browser = webdriver.Chrome()

# Open the URL
browser.get(url)

# Wait for the page to load completely
browser.implicitly_wait(10)

all_titles = []
# Click the 'Load More' button 10 times
for _ in range(10):
    try:
        # Find all elements with the class 'c-gallery-vertical-album__title'
        titles_temp = browser.find_elements(By.CLASS_NAME, 'c-gallery-vertical-album__title')
        for title in titles_temp:
            title_text = title.text
            if title_text not in all_titles:
                print(title_text)
                all_titles.append(title_text)
                
        # Check if "Load More" button exists
        load_more_buttons = browser.find_elements(By.XPATH, '//a[text()="Load More"]')
        if not load_more_buttons:
            print("No more 'Load More' button found.")
            break  # break the loop if the button is not found
        
        # Click the "Load More" button
        load_more_button = load_more_buttons[0]
        load_more_button.click()
        time.sleep(2)  # wait for 2 seconds after each click for the content to load
    except Exception as e:
        print(f"An error occurred: {e}")
        break  # break the loop if there's an error

# Close the browser window
browser.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import TimeoutException
with open('rollingstone_top_130.txt', encoding='utf-8') as file:
    all_titles = [line.strip() for line in file.readlines()]

def get_youtube_url(song_name):
    # Initialize a new instance of Chrome
    browser = webdriver.Chrome()

    # Open YouTube
    browser.get('https://www.youtube.com')

    # Find the search box and enter the song name
    search_box = browser.find_element(By.NAME, 'search_query')
    search_box.send_keys(song_name)
    search_box.send_keys(Keys.RETURN)
    time.sleep(3)
    try:
        # Wait for the video titles to load
        WebDriverWait(browser, 10).until(
            EC.presence_of_element_located((By.ID, 'video-title'))
        )

        # Find the <a> element by ID
        a_elements = browser.find_elements(By.ID, 'video-title')
        for a_element in a_elements:
            # Print the outer HTML of the <a> element
            outer_html = a_element.get_attribute('outerHTML')
           
            if song_name.lower() in outer_html.lower():
                print(f"The outer HTML of the <a> element contains the search query: {song_name}")
            outer_html = a_element.get_attribute('innerHTML')
            if song_name.lower() in outer_html.lower():
                print(f"The Inner HTML of the <a> element contains the search query: {song_name}")
            # Alternatively, print only the href attribute
            href = a_element.get_attribute('href')
            if href !=None:
                break
                print(f"href attribute of the <a> element: {href}")

        # Use XPath to find the video based on its text content and click it
        # video = browser.find_element(By.XPATH, f"//a[@id='video-title']//yt-formatted-string[contains(text(),'{song_name}')]")
        # video.click()

        # browser.execute_script("document.getElementById('video-title').click();")
     
      
        # Wait for the new page to load
        # time.sleep(3)  # Adjust the sleep time as needed

        # Get the current URL
        url = href
 
    except Exception as e:
        print(f"An error occurred: {e}")
        url = None

    # Close the browser
    browser.quit()

    return url

# # Example usage
# song_name = 'Bohemian Rhapsody Queen'  # Replace with the name of the song
# url = get_youtube_url(song_name)
# print(url)


#Getting URLs for all titles in the all_titles list
all_urls = []
for title in all_titles:
    url = get_youtube_url(title)
    all_urls.append(url)
    print(f"URL for '{title}': {url}")

# Now all_urls list contains the YouTube URLs for all titles in the all_titles list
```

```python
import random
import webbrowser
import time
from pytube import YouTube
import keyboard

def get_video_duration(url):
    # Create a YouTube object
    yt = YouTube(url)

    # Get the duration of the video
    duration_seconds = yt.length  # duration in seconds

    # Convert the duration to minutes and seconds
    #minutes, seconds = divmod(duration_seconds, 60)

    return duration_seconds

with open('URLs_top_500.txt', encoding='utf-8') as file:
    song_urls = [line.strip() for line in file.readlines()]

```

```python
while True:
    # Choose a random song URL from the list
    random_song = random.choice(song_urls)

    
    # Extract the actual URL from the string
    url = random_song.split(' ')[-1]
    
    # Open the URL in the default web browser
    webbrowser.open(url)
    
    # Wait for some time before playing the next song
    # Adjust the sleep time based on the average length of the songs
    duration = get_video_duration(url)
    
      # Check for 'q' press in smaller intervals until the entire duration is covered
    check_interval = 5  # seconds
    total_time_checked = 0

    while total_time_checked < duration:
        if keyboard.is_pressed('q'):
            break
        time.sleep(check_interval)
        total_time_checked += check_interval
    
    if keyboard.is_pressed('q'):
        break
```

```python
exit_loop = False

while True and not exit_loop:
    # Choose a random song URL from the list
    random_song = random.choice(song_urls)

    # Extract the actual URL from the string
    url = random_song.split(' ')[-1]
    
    # Open the URL in the default web browser
    webbrowser.open(url)
    
    # Get the video duration
    duration = get_video_duration(url)
    
    # Remove the played song from the list
    song_urls.remove(random_song)

    # Check for 'q' press in smaller intervals until the entire duration is covered
    check_interval = 0.1  # seconds
    total_time_checked = 0
    
    while total_time_checked < duration:
        if keyboard.is_pressed('q'):
            exit_loop = True
            print(exit_loop)
            break
        time.sleep(check_interval)
        total_time_checked += check_interval
```
